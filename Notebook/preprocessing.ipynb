```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/crime_sample_500k.csv', low_memory=False)
df['Date'] = pd.to_datetime(df['Date'])
print(df.shape)
```

```python
df.isnull().sum()
```

```python
# Temporal features
df['Hour'] = df['Date'].dt.hour
df['Day_of_Week'] = df['Date'].dt.day_name()
df['Month'] = df['Date'].dt.month
df['Is_Weekend'] = df['Date'].dt.dayofweek >= 5
df['Season'] = df['Month'].map({
    12:'Winter',1:'Winter',2:'Winter',
    3:'Spring',4:'Spring',5:'Spring',
    6:'Summer',7:'Summer',8:'Summer',
    9:'Fall',10:'Fall',11:'Fall'
})

# Crime severity score
severity = {
    'HOMICIDE':10,'CRIM SEXUAL ASSAULT':9,'KIDNAPPING':9,
    'ARSON':8,'ROBBERY':8,'ASSAULT':7,'BATTERY':6,
    'WEAPONS VIOLATION':7,'BURGLARY':6,'MOTOR VEHICLE THEFT':5,
    'THEFT':4,'NARCOTICS':4,'CRIMINAL DAMAGE':3,
    'CRIMINAL TRESPASS':2,'OTHER OFFENSE':2
}
df['Crime_Severity'] = df['Primary Type'].map(severity).fillna(1)

print(df[['Hour','Day_of_Week','Season','Crime_Severity']].head())
print(f"\nNew columns added: {['Hour','Day_of_Week','Month','Is_Weekend','Season','Crime_Severity']}")
```

```python
df.to_csv('../data/crime_features.csv', index=False)
print(f"Saved → data/crime_features.csv  |  Shape: {df.shape}")
```
